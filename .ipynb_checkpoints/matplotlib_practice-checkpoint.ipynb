```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
plt.xlim(0, 5)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot.
# plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')

# # Create labels for the x and y axes.
# plt.xlabel("Date")
# plt.ylabel("Fare($)")
# Change the line color to green.

# Change the marker to a diamond.

# Define ax as plot
fig, ax = plt.subplots()
# Define / say how you want data to display
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# # Add a legend for the city of Boston.
ax.legend()
# # Add a title and axes labels.
ax.set_title("PyBer SkillDrill")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# # Add a y-axis limit.
ax.set_ylim(0, 45)
# # Add grid lines.
ax.grid()
```

```python
y_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
x_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.scatter(x_axis,y_axis, c='red')
plt.xlabel ('Fare($)')
plt.ylabel ('Date')
plt.title ('PyBer Fare by Month')
plt.legend('Chicago')
plt.xlim(0, 50)
plt.gca().invert_yaxis()
plt.show()
```

```python
y_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
x_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*5)
plt.scatter(x_axis,y_axis, c='skyblue',marker="*", linewidth=2, label='Boston')
plt.ylabel ('Fare($)')
plt.xlabel ('Date')
plt.title ('PyBer Fare by Month')
plt.legend()
plt.xlim(0, 50)
plt.gca().invert_yaxis()
plt.show()
```
